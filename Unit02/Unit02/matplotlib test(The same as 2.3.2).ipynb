```python
import matplotlib.pyplot as plt

days = range(1,7)
celsius = [25.6, 23.2, 18.5, 28.3, 26.5, 30.5]
plt.plot(days, celsius, "r-o")   # "r-o" & "o-r" 相同
plt.grid(True)   #顯示表格
plt.xlabel("days")
plt.ylabel("Degree")
plt.title("Temperature of the week ")
plt.legend (loc=1)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")

labels = ["Python", "C++", "Java", "JS", "C", "C#"]
index = np.arange(len(labels))
ratings = [5.16, 5.73, 14.99, 3.17, 11.86, 4.45]
change = [1.12, 0.3, -1.69, 0.29, 3.41, -0.45]
```

```python
plt.bar(index,ratings)   
plt.xticks(index,labels)   # x軸要給什麼標籤
plt.ylabel("Rating")
plt.title("Programming rating")
plt.show()
```

```python
plt.barh(index,change)   
plt.yticks(index,labels)   # x軸要給什麼標籤
plt.xlabel("Change")

plt.title("Programming Change rate")
plt.show()
```

```python
x = [21, 42, 23, 4, 5, 26, 77, 88, 9, 10, 31, 32, 33,
     34, 35, 36, 37, 18, 49, 50, 100]
num_bins = 5   # 多少區間分割
n, bins, patches = plt.hist(x,num_bins)   # n = 每個區間有幾個值 # bins = 每個長條圖的分隔線 #
print(n)
print(bins)
print(patches)  # 每個bins中所含的資料，他會是個list的形式
plt.show()
```

```python
# Pie plots

labels = ["Python", "C++", "Java", "JS", "C", "C#"]
ratings2 = [5, 6, 15, 3, 12, 4]

plt.pie(ratings2, labels = labels)
plt.title("Programing using rate")
plt.axis("equal")
plt.show()
```

```python
# 子圖表(subplots)
x = np.linspace(0, 10, 50)  #linspace(起始範圍 , 結束範圍, 取幾個數) 指定的區間内生成均匀間隔的數字
sinus = np.sin(x)
cosinus = np.cos(x)
plt.subplot(2, 1, 1) # 這邊分成兩區塊 第三個數字代表先畫哪格圖     #plt.subplot(num_rows, num_columns, plot_num)
plt.plot(x, sinus, "r-o")
plt.subplot(2, 1, 2)
plt.plot(x, cosinus, "g--")
plt.show()
```

```python
sinhs = np.sinh(x)  #雙曲函數
fig , ax = plt.subplots()
ax.plot(x, sinus, "r-o")
ax2 = ax.twinx() #twins 共用
ax2.plot(x, sinhs, "g--")
plt.show()
```
